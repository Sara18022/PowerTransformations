# Glavne karakteristike normalne raspodele

**Raspodela podataka** predstavlja način na koji su vrednosti određene promenljive koju posmatramo raspoređene unutar skupa podataka, pokazujući učestalost (frekvenciju) pojavljivanja svake vrednosti ili grupe vrednosti u tom skupu. 

**Normalna raspodela**, poznata i kao Gaussova raspodela, jedna je od najvažnijih i najčešće korišćenih raspodela u statistici. Ova raspodela ima oblik zvonaste krive koja je simetrična oko svoje srednje vrednosti. Naziva se "normalna" jer mnogi biološki i prirodni fenomeni, poput visine, težine i nivoa šećera u krvi, prate ovaj obrazac.

Ključni pojmovi:

 - **Srednja vrednost (μ)**: najviša tačka na krivoj raspodele, podaci su simetrično raspoređeni oko ove vrednosti.
 - **Standardna devijacija (σ)**: mera disperzije ili rasipanja podataka oko srednje vrednosti. Manja standardna devijacija znači da su podaci bliže srednjoj vrednosti, dok veća standardna devijacija ukazuje na veće rasipanje podataka.
 


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Postavljamo različite srednje vrednosti i standardne devijacije
params = [
    {"mu": 0, "sigma": 1, "label": "μ=0, σ=1"},
    {"mu": 0, "sigma": 2, "label": "μ=0, σ=2"},
    {"mu": 2, "sigma": 1, "label": "μ=2, σ=1"},
    {"mu": -2, "sigma": 0.5, "label": "μ=-2, σ=0.5"},
]

plt.figure(figsize=(12, 8))

# Generisemo podatke za svaku kombinaciju mu i sigma i crtamo na istom grafiku
for param in params:
    mu = param["mu"]
    sigma = param["sigma"]
    data = np.random.normal(mu, sigma, 1000)
    sns.kdeplot(data, label=param["label"], linewidth=2)

# Dodavanje naslova, etiketa i legende
plt.title('Poređenje Normalnih Raspodela za Različite Vrednosti μ i σ')
plt.legend(title="Parametri")
plt.show()
```

Karakteristike normalne raspodele:
   1) Podaci su simetrično raspoređeni oko srednje vrednosti (μ).
   
   2) Srednja vrednost, medijana (srednja vrednost skupa podataka kada su svi podaci poređani po veličini) i mod (vrednost koja se najčešće pojavljuje u skupu podataka) su iste vrednosti i nalaze se u centru krive, najveći broj podataka gravitira ka sredini raspodele.
   
   3) Manja standardna devijacija znači užu, višu krivu, dok veća standardna devijacija ukazuje na širu, nižu krivu.
   
   4) Repovi normalne raspodele su beskonačno dugi i tanki su, što znači da ekstremne vrednosti (veoma male ili veoma velike) imaju nisku, ali nenultu verovatnoću pojavljivanja.
