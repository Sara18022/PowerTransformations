```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from scipy.stats import boxcox
```

```python
plt.figure(figsize = (8, 8))
data = np.random.beta(1, 3, 5000)
sns.histplot(data, kde=True)
plt.show()
```

```python
tdata = boxcox(data)[0]
plt.figure(figsize = (8, 8))
sns.histplot(tdata, kde=True) 
plt.show()
```
